# Deep Learning

* You can consider Deep Learning of an automated version of Machine Learning.
* This automation is achieved by the inspiration from the neural system of from our brain.

* You might know that our brain consists of billions of neurons which are interconnected to each other forming a network called neual network.

* A neuron is so simple, that it takes some inputs and decides whether to be active or not, but it is the interconnects that makes things complicated and powerful. 

* Inspiring from this, we invented **Artificial** Neural Networks that mimic the same.

* We have neurons (yellow balls) and interconnects(yellow lines). But how do we make them work?

* So, each neuron takes inputs from the interconects that it is connected to, decide some importance to each input, sum them, and output.
* The importance to different input is represented by the weights, we give to the interconnects.

* This arrangement can do a lots of powerful things which we will see.

* There are many types of these networks,

## 1)Feed Forward Neural Network

* This is a naive version of the neural networks which we use for general Machine Learning problems, like solving regression, classification problems based on numeric and categorical data.
* Image and other data related problems can be solved but not preferred.

## 2) Convolutional Neural Network

* These networks are specialized to solve image based problems. If any problem involves image, then you have to look at CNN.

**Where are they used??**
* The property of CNN is to identify a feature irrespective of its location in the features.
* Examples include Images, words in a sentence, ect.
* They are also used when we want to parallelize same computation.

## 3) Recurrent Neural Networks

* These networks are used to deal with sequential or time based data.
* It is mostly used in NLP applications and handle temporal data.

# Activation Functions
* If you consider a network without activation function, it is just linear like the regression problem.

* We add activation functions at the output end of each neuron, to make complex representations,which is why we use neural networks.
* So, In a every neuron will definitely have a activation function except at the final layer if it a regression problem.

## 1) Sigmoid Activation
* This is a basic activation function which squeezes the output between 0 and 1.


## 2) Tanh
* This activation function squeezes output between -1 and 1.
* This is preferred to sigmoid because it has 0 mean and proves to converge well in most cases.
* It is mostly used in RNN's.


## 3) Softmax
* This function is used in last layer of a multiclass classification problem as its output multiple probabilities which sums to one.
* It is computationally expensive, and if there are more classes, sigmoid or tanh on individual class is preffered.


## 4) ReLU
* This function is the most used activation function by default.
* It is computationally cheap, solves vanishing gradient problems which we will see later.

* It has variants like Leaky ReLu,PRELu,ELU, Swish,Maxout which help in handling negative values, since Relu has zero gradient over negative values.

**Best Practices** 
* For intermediate layers, start with ReLu.
* If not, Update to Leaky ReLu.
* Else try its variants.
